### General

```python
# Import Libraries
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
```

```python
# Read Sequences
seqA = SeqIO.read('./Exercises/a.fasta', 'fasta')
seqB = SeqIO.read('./Exercises/b.fasta', 'fasta')

# inget diubah ke Seq
seqA = seqA.seq
seqB = seqB.seq

print(seqA)
print(seqB)
```

### 1. Sequence Manipulation

```python
# a. Find and display the total length of Sequence A and B.
print(f"Total length of Seq A: {len(seqA)}")
print(f"Total length of Seq B: {len(seqB)}")

# b. Find and display which index the first time codon (TAC) appears 
#    in both Sequence A and B.
print(f"First index TAC of Seq A found on index {seqA.find('TAC')}")
print(f"First index TAC of Seq B found on index {seqB.find('TAC')}")


# c. Take the first 10 nucleotides from Sequence A, and the last 15 nucleotides 
#    from Sequence B. Then combine the sequences and reverse it to a new sequence, 
#    Sequence C.
seqA_sliced = seqA[:10]
seqB_sliced = seqB[-15:]
seqC = (seqA_sliced + seqB_sliced)[::-1]
print(f"Seq C : {seqC}")
```

### 2. Sequence Analysis and Plotting

```python
# jangan lupa imort librarynya buat plotting
import matplotlib.pyplot as plt
from collections import Counter

# a. Determine and display the nucleotides base (A, C, G, T) frequency 
#    for Sequence A, B, and C with a chart using matplotlib library.
seqA_freq = Counter(seqA)
seqB_freq = Counter(seqB)
seqC_freq = Counter(seqC)

plt.title("Nucleotide base frequency for Seq A")
plt.bar(seqA_freq.keys(), seqA_freq.values())
plt.show()

plt.title("Nucleotide base frequency for Seq B")
plt.bar(seqB_freq.keys(), seqB_freq.values())
plt.show()

plt.title("Nucleotide base frequency for Seq C")
plt.bar(seqC_freq.keys(), seqC_freq.values())
plt.show()
```

```python
# b. Determine and display the GC & AT content percentage for Sequence A, B, and C.
# from Bio.SeqUtils import gc_fractions
# seqA_GC = gc_fraction(seqA) * 100
# seqB_GC = gc_fraction(seqB) * 100
# seqC_GC = gc_fraction(seqC) * 100

from Bio.SeqUtils import GC
seqA_GC = GC(seqA)
seqB_GC = GC(seqB)
seqC_GC = GC(seqC)

seqA_AT = 100 - seqA_GC
seqB_AT = 100 - seqB_GC
seqC_AT = 100 - seqC_GC

print(f"SeqA GC Percentage: {seqA_GC:.2f}%, AT Percentage: {seqA_AT:.2f}%")
print(f"SeqB GC Percentage: {seqB_GC:.2f}%, AT Percentage: {seqB_AT:.2f}%")
print(f"SeqC GC Percentage: {seqC_GC:.2f}%, AT Percentage: {seqC_AT:.2f}%")


# c. Determine and display the melting temperature for Sequence A, B,
#    and C (using built-inWallace Method).
from Bio.SeqUtils import MeltingTemp as mt
print(f"Melting temperature for SeqA: {mt.Tm_Wallace(seqA)}")
print(f"Melting temperature for SeqB: {mt.Tm_Wallace(seqB)}")
print(f"Melting temperature for SeqC: {mt.Tm_Wallace(seqC)}")


# d. Determine and display the molecular weight for Sequence A, B, and C.
from Bio.SeqUtils import molecular_weight as mw
print(f"Molecular Weight for SeqA: {mw(seqA)}")
print(f"Molecular Weight for SeqB: {mw(seqB)}")
print(f"Molecular Weight for SeqC: {mw(seqC)}")
```

### 3. Protein Synthesis

```python
# a. Transcribe and display the DNA Sequence A, B, and C into mRNA Sequences.
mRNA_seqA = seqA.transcribe()
mRNA_seqB = seqB.transcribe()
mRNA_seqC = seqC.transcribe()

print(f"mRNA of SeqA: {mRNA_seqA}")
print(f"mRNA of SeqB: {mRNA_seqB}")
print(f"mRNA of SeqC: {mRNA_seqC}")


# b. Translate and display the mRNA Sequence A, B, and C into amino acids sequences.
#    (Clue: for translate, use the mRNA sequences that already transcribed on the first step)
amino_seqA = mRNA_seqA.translate()
amino_seqB = mRNA_seqB.translate()
amino_seqC = mRNA_seqC.translate()

print(f"Amino acids of SeqA: {amino_seqA}")
print(f"Amino acids of SeqB: {amino_seqB}")
print(f"Amino acids of SeqC: {amino_seqC}")
```

### 4. Sequence Alignment & Similarities

```python
# a. Perform local alignment pairwise between Sequence A and Sequence CGCTACTA
#    GCATGCGCAT and display the alignments score. (Clue: Use localxx)
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

alignment_seqA = pairwise2.align.localxx(seqA, 'CGCTACTAGCATGCGCAT')

# Format
top_alignment = alignment_seqA[0]
print(format_alignment(*top_alignment))
# Score
print(f"Alignment of SeqA Score: {top_alignment[2]}")


# b. Perform global alignment pairwise between Sequence B and Sequence CAGTGCG
#    CTAGCAGCTACGTGTG and display the alignments score. (Clue: Use globalxx)
alignment_seqB = pairwise2.align.globalxx(seqB, 'CAGTGCGCTAGCAGCTACGTGTG')
# Format
top_alignment = alignment_seqB[0]
print(format_alignment(*top_alignment))
# Score
print(f"Alignment of SeqB Score: {top_alignment[2]}")
```
