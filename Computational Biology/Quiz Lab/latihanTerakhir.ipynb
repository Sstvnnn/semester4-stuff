### General

```python
# Libraries
import Bio
from Bio import SeqIO
from Bio.Seq import Seq
```

```python
# Read Fasta
seqA = SeqIO.read('./Exercises/a.fasta', 'fasta')
seqB = SeqIO.read('./Exercises/b.fasta', 'fasta')

seqA = seqA.seq
seqB = seqB.seq

print(f"SeqA: {seqA}")
print(f"SeqB: {seqB}")
```

### 1. Sequence Manipulation

```python
# a. Total Length
print(f"Total length of SeqA: {len(seqA)}")
print(f"Total length of SeqB: {len(seqB)}")

# b. First Index codon "TAC" Appears
print(f"First Index codon 'TAC' Appears of SeqA: {seqA.find('TAC')}")
print(f"First Index codon 'TAC' Appears of SeqB: {seqB.find('TAC')}")

# c. Sequence C (Slicing)
seqA_sliced = seqA[:10]
seqB_sliced = seqB[-15:]
seqC = (seqA_sliced+seqB_sliced)[::-1]
print(f"SeqC: {seqC}")
```

### 2. Sequence Analysis and Plotting

```python
# a. Nucleotide Base Frequency
import matplotlib.pyplot as plt
from collections import Counter

seqA_freq = Counter(seqA)
seqB_freq = Counter(seqB)
seqC_freq = Counter(seqC)

plt.title("Nucleotide Base Frequency of Sequence A")
plt.bar(seqA_freq.keys(), seqA_freq.values())
plt.show()

plt.title("Nucleotide Base Frequency of Sequence B")
plt.bar(seqB_freq.keys(), seqB_freq.values())
plt.show()

plt.title("Nucleotide Base Frequency of Sequence C")
plt.bar(seqC_freq.keys(), seqC_freq.values())
plt.show()
```

```python
#  b. GC AT Content Percentage
from Bio.SeqUtils import GC

seqA_GC = GC(seqA)
seqB_GC = GC(seqB)
seqC_GC = GC(seqC)

seqA_AT = 100 - seqA_GC
seqB_AT = 100 - seqB_GC
seqC_AT = 100 - seqC_GC

print(f"SeqA GC Percentage: {seqA_GC:.2f}%, AT Percentage: {seqA_AT:.2f}%")
print(f"SeqB GC Percentage: {seqB_GC:.2f}%, AT Percentage: {seqB_AT:.2f}%")
print(f"SeqC GC Percentage: {seqC_GC:.2f}%, AT Percentage: {seqC_AT:.2f}%")

# c. Melting Temperature
from Bio.SeqUtils import MeltingTemp as MT
print(f"SeqA Melting Temperature: {MT.Tm_Wallace(seqA)}")
print(f"SeqB Melting Temperature: {MT.Tm_Wallace(seqB)}")
print(f"SeqC Melting Temperature: {MT.Tm_Wallace(seqC)}")

# d. Molecular Weight
from Bio.SeqUtils import molecular_weight as MW
print(f"SeqA Molecular Weight: {MW(seqA)}")
print(f"SeqB Molecular Weight: {MW(seqB)}")
print(f"SeqC Molecular Weight: {MW(seqC)}")
```

### 3. DNA & mRNA Protein Synthesis

```python
# a. Transcribe
seqA_mRNA = seqA.transcribe()
seqB_mRNA = seqB.transcribe()
seqC_mRNA = seqC.transcribe()

print(f"mRNA of SeqA:{seqA_mRNA}")
print(f"mRNA of SeqB:{seqB_mRNA}")
print(f"mRNA of SeqC:{seqC_mRNA}")

# b. Translate
seqA_amino = seqA_mRNA.translate()
seqB_amino = seqB_mRNA.translate()
seqC_amino = seqC_mRNA.translate()

print(f"Amino acid of SeqA: {seqA_amino}")
print(f"Amino acid of SeqB: {seqB_amino}")
print(f"Amino acid of SeqC: {seqC_amino}")
```

### 4. Sequence Alignment & Similarities

```python
# a. Local Alignment Pairwise between Seq A and Seq CGCTACTAGCATGCGCAT
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

seqA_alignment = pairwise2.align.localxx(seqA, 'CGCTACTAGCATGCGCAT')
top_alignment = seqA_alignment[0]
print(format_alignment(*top_alignment))
print(f"SeqA Alignment Score: {top_alignment[2]}")

# b. Global Alignment Pairwise between Seq B and Seq "CAGTGCGCTAGCAGCTACGTGTG"
seqB_alignment = pairwise2.align.globalxx(seqB, 'CAGTGCGCTAGCAGCTACGTGTG')
top_alignment = seqB_alignment[0]
print(format_alignment(*top_alignment))
print(f"SeqA Alignment Score: {top_alignment[2]}")
```

```python
# c. Levenshtein Distance between Seq A and Seq B'
from Levenshtein import distance
print(f"Levenshtein Distance between SeqA and SeqB: {distance(seqA, seqB)}")
```
